### Estilos

Las versiones más recientes de Matplotlib permiten la aplicación de "estilos" que mejoran la apariencia de las gráficas creadas o que las modifican de forma tal que las hacen más amigables en ciertos entornos.

Podemos comprobar los estilos disponibles con la instrucción **plt.style.available**:

```python
plt.style.available
```

Para usar un estilo, por ejemplo, "seaborn", basta ejecutar la siguiente instrucción:

plt.style.use("seaborn")

A continuación se muestran dos gráficas en todos los estilos disponibles. Puede apreciarse la mejora que supone el uso de estilos con respecto al estilo "classic" propio de Matplotlib. Obsérvese también que hay estilos que -al menos para las gráficas y elementos mostrados- no aparentan mostrar diferencias:

```python
styles = plt.style.available
```

```python
y1 = generator.standard_normal(100).cumsum()
y2 = generator.standard_normal(100) * 10
```

```python
def showGraphs(style):
    plt.style.use(style)
    fig, ax = plt.subplots(1, 2)
    fig.set_size_inches(10, 3)
    fig.suptitle('Estilo ' + style,
                 fontsize = 16,
                 backgroundcolor = "CornflowerBlue",
                 color = "w",
                 position = (0.5, 1.1)
                )
    ax[0].plot(y1)
    ax[1].hist(y2)
    plt.show()
```

### Mapas de color

Matplotlib ofrece un conjunto de mapas de color predefinidos. Éstos son diccionarios de colores que "mapean" los valores representados en nuestras gráficas a ciertos conjunto de colores. La idea detrás de los mapas de colores es encontrar una buena representación en, por ejemplo, espacios tridimensionales, colores capaces de transmitir la sensación de profundidad que los colores lisos no transmiten correctamente. La elección correcta de un mapa de color puede cambiar radicalmente la forma en la que percibimos una gráfica. Puedes encontrar documentación sobre estos mapas de colores [en esta página de matplotlib](https://matplotlib.org/stable/users/explain/colors/colormaps.html).

Las funciones gráficas que soportan mapas de color aceptan el parámetro **cmap**. Éste deberá tomar un valor correspondiente a un mapa de color existente, como "viridis" o "BrBG".

Ten en cuenta que los estilos pueden interferir en la forma en la que se aplican los mapas de color.

Podemos mostrar los mapas de color disponibles por defecto usando la función **plt.colormaps**:

```python
print(plt.colormaps())
```
