Aun cuando Matplotlib fue diseñado en un principio como software para la generación de gráficos bidimensionales, con el tiempo se le han ido añadiendo herramientas que, si bien limitadas, permiten la generación de gráficos en 3D.

Para habilitar la creación de este tipo de gráficos basta crear un conjunto de ejes indicando que la proyección es de tipo "*3d*". Para esto podemos recurrir, por ejemplo, al métido *.add_subplot()* (no todas las formas de creación de conjunto de ejes permiten especificar el tipo de proyección de forma sencilla):

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
plt.show()
```

A partir de aquí, basta con utilizar los métodos de los conjuntos de ejes que generan los gráficos en 3D.

### Superficies

Una vez creado el conjunto de ejes 3D, mostrar una superficie tridimensional en estos ejes es relativamente sencillo usando el método del conjunto de ejes [plot_surface](https://matplotlib.org/stable/plot_types/3D/surface3d_simple.html). Este método recibe como argumentos de entrada arrays 2D con los valores $x$, $y$ y $z$ de la misma manera que vimos para la función *plt.contour()*. Por ejemplo, partamos del siguiente conjunto de puntos:

```python
x = np.arange(-10, 10, 0.25)
y = np.arange(-10, 10, 0.25)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
```

Ahora no queda más que crear la figura, el conjunto de ejes y ejecutar el método **.plot_surface()** mencionado. En el ejemplo de la siguiente figura, hemos asignado a la gráfica un mapa de color y creado una barra de color:

```python
fig = plt.figure(figsize = (12, 6))
ax = fig.add_subplot(projection='3d')
surface = ax.plot_surface(X, Y, Z, cmap = "coolwarm")
fig.colorbar(surface)
plt.show()
```

### Wireframes

La generación de "wireframes" o gráficos de mallas es extremadamente simple una vez que sabemos generar una superficie. El método es [plot_wireframe](https://matplotlib.org/stable/plot_types/3D/wire3d_simple.html) y recibe como primeros tres argumentos las matrices $X$, $Y$ y $Z$ que hemos visto para el método *.plot_surface()*. En este caso, el método no es compatible con los mapas de color, por lo que no es posible asignarle uno ni crear una barra de color que acompañe al gráfico:

```python
fig = plt.figure(figsize = (12, 6))
ax = fig.add_subplot(projection='3d')
surface = ax.plot_wireframe(X, Y, Z)
plt.show()
```

Con los parámetros **rcount** y **ccount** (cuyo valor por defecto es 50) es posible especificar el número de muestras de los datos a utilizar en cada dirección. En el siguiente ejemplo utilizamos apenas 25 muestras por dirección:

```python
fig = plt.figure(figsize = (12, 6))
ax = fig.add_subplot(projection='3d')
surface = ax.plot_wireframe(X, Y, Z, rcount = 25, ccount = 25)
plt.show()
```

### Curvas de nivel

El método *.contour3D()* nos permite crear gráficos de curvas de nivel en 3D. En su configuración más básica simplemente recibe como argumentos los arrays $X$, $Y$ y $Z$:

```python
fig = plt.figure(figsize = (12, 6))
ax = fig.add_subplot(projection='3d')
surface = ax.contour3D(X, Y, Z, cmap = "hot")
fig.colorbar(surface)
plt.show()
```

Por defecto, las curvas de nivel se crean paralelas al eje $x$ (es decir, en torno al eje $y$), pero el parámetro **zdir** permite modificar el eje de referencia. Además, el parámetro **levels** nos permite escoger el número de niveles a mostrar:

```python
fig = plt.figure(figsize = (12, 6))
ax = fig.add_subplot(projection='3d')
surface = ax.contour3D(X, Y, Z, cmap = "hot", zdir = "x", levels = 20)
fig.colorbar(surface)

plt.show()
```

### Gráficos de dispersión

También podemos crear gráficos de dispersión fácilmente con el método [scatter3D](https://matplotlib.org/stable/gallery/mplot3d/scatter3d.html). Una vez más, los tres argumentos principales son los arrays $X$, $Y$ y $Z$ conteniendo la información sobre los puntos a mostrar. En el siguiente ejemplo generamos dichos datos de forma aleatoria:

```python
n = 4
fig = plt.figure(figsize = (12, 6))
ax = fig.add_subplot(projection='3d')
x = np.random.normal(0, (n + 1) * 3, 500)
y = np.random.normal(0, (n + 1) * 3, 500)
z = np.random.normal(0, (n + 1) * 3, 500)
scatter = ax.scatter3D(x, y, z)
plt.show()
```

Podemos, por supuesto, superponer diferentes grupos de puntos y asignar una etiqueta a cada uno, tal y como hacíamos con los gráficos de dispersión 2D:

```python
fig = plt.figure(figsize = (12, 6))
ax = fig.add_subplot(projection='3d')
for n in range(3):
    x = np.random.normal(0, (n + 1) * 3, 100)
    y = np.random.normal(0, (n + 1) * 3, 100)
    z = np.random.normal(0, (n + 1) * 3, 100)
    scatter = ax.scatter3D(x, y, z, label = n)
plt.legend()
plt.show()
```
