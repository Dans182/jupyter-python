```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Pandas permite la creación de gráficos mediante funciones integradas, facilitando de esta forma la exploración y comprensión de los conjuntos de datos. En realidad se basa en Matplotlib para la generación de los gráficos, pero simplifica su uso mediante interfaces más sencillas y directas.

La funcionalidad básica está accesible a través del método **.plot** de series y dataframes. Este método nos permite crear diferentes gráficas a partir de los datos contenidos en la estructura Pandas. Adicionalmente, dataframes y series tienen métodos para la creación de gráficas que, en esencia, son atajos para ejecutar los mismos gráficos a través del método *.plot*.

Otras funciones más avanzadas están disponibles en la sublibrería **pandas.plotting**.

Carguemos algunos datasets de Seaborn:

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
flights = sns.load_dataset("flights").groupby("year")["passengers"].sum().to_frame().reset_index()
flights.head(2)
```

```python
iris = sns.load_dataset("iris")
tips.head(2)
```

# Funcionalidad básica

El método *.plot* de dataframes y series incluye el parámetro **kind** que nos permite especificar el tipo de gráfico a crear:

* **"line"**: Gráficos de líneas (valor por defecto)
* **"bar"**: Gráficos de columnas
* **"barh"**: Gráficos de barras
* **"hist"**: Histogramas
* **"box"**: Gráficos de caja (boxplots)
* **"scatter"**: Gráficos de dispersión (scatter plots)
* **"pie"**: Gráficos circulares (pie charts)
* **"kde"** o **"density"**: Estimaciones de densidad Kernel (kde)
* **"area"**: Gráficos de áreas
* **"hexbin"**: Gráfico de hexágonos (hexbin plots)

### Gráfico de líneas

Por ejemplo, podríamos crear un gráfico de líneas con:

```python
flights.plot(
    kind = "line",
    figsize = (8, 2.5),
    x = "year",
    y = "passengers",
    marker = 'o',
    title = "Passengers by year",
    xlabel = "Year",
    ylabel = "Passengers",
    grid = True
)
plt.show()
```

El valor por defecto del parámetro *kind* es "*line*", de forma que no hubiese sido necesario incluirlo.

Comprobamos que el método *.plot* acepta parámetros equivalentes a los nombres de funciones existentes en Matplotlib para dar formato a ejes, etc.

La misma gráfica está disponible como método de *.plot*:

```python
flights.plot.line(
    figsize = (8, 2.5),
    x = "year",
    y = "passengers",
    marker = 'o',
    title = "Passengers by year",
    xlabel = "Year",
    ylabel = "Passengers",
    grid = True
)
plt.show()
```

### Gráfico de columnas

Accesible a través de *.plot*:

```python
flights.plot(
    kind = "bar",
    figsize = (8, 2.5),
    x = "year",
    y = "passengers",
    color = "salmon",
    legend = False,
    title = "Passengers by year",
    xlabel = "Year",
    ylabel = "Passengers",
    grid = True
)
plt.show()
```

O a través del metodo **.plot.bar()**:

```python
flights.plot.bar(
    figsize = (8, 2.5),
    x = "year",
    y = "passengers",
    color = "salmon",
    legend = False,
    title = "Passengers by year",
    xlabel = "Year",
    ylabel = "Passengers",
    grid = True
)
plt.show()
```

### Gráfico de barras

```python
flights.plot(
    kind = "barh",
    figsize = (8, 3.5),
    x = "year",
    y = "passengers",
    color = "Teal",
    legend = False,
    title = "Passengers by year",
    xlabel = "Passengers",
    ylabel = "Year",
    grid = True
)
plt.show()
```

```python
flights.plot.barh(
    figsize = (8, 3.5),
    x = "year",
    y = "passengers",
    color = "Teal",
    legend = False,
    title = "Passengers by year",
    xlabel = "Passengers",
    ylabel = "Year",
    grid = True
)
plt.show()
```

### Gráfico circular

Agrupamos los importes cobrados por día de la semana:

```python
total_bill_by_day = tips.groupby(by = "day", observed = False)["total_bill"].sum()
total_bill_by_day
```

```python
total_bill_by_day.plot(
    kind = "pie",
    figsize = (6, 4),
    y = total_bill_by_day,
    labels = total_bill_by_day.index,
    autopct='%1.1f%%',
    title = "Total bill by day",
    ylabel = ""
)
plt.show()
```

```python
total_bill_by_day.plot.pie(
    figsize = (6, 4),
    y = total_bill_by_day,
    labels = total_bill_by_day.index,
    autopct='%1.1f%%',
    title = "Total bill by day",
    ylabel = ""
)
plt.show()
```

# Funcionalidad adicional

Además de las funciones que hemos visto, Pandas nos ofrece herramientas adicionales:

### subplots

El asignar al parámetro **subplots** el valor *True* provoca la generación de una gráfica para cada una de las columnas involucradas:

```python
iris.plot.hist(
    figsize = (5, 6),
    bins=20,
    alpha=0.7,
    xlabel = "cm",
    ylabel = "Frequency",
    grid = True,
    subplots = True
)
plt.show()
```

### layout

Y, en el caso de que estemos generando más de una gráfica, podemos controlar la disposición de éstas con el parámetro **layout**. Así, si queremos repetir la gráfica anterior mostrando los cuatro histogramas en una única fila, podemos hacerlo del siguiente modo:

```python
iris.plot.hist(
    figsize = (12, 3),
    bins=20,
    alpha=0.7,
    xlabel = "cm",
    ylabel = "Frequency",
    grid = True,
    subplots = True,
    layout = (1, 4)
)
plt.show()
```

### ax

Si generamos la figura usando funciones nativas de Matplotlib e incluimos en ella más de un eje, también podemos especificar en qué eje queremos que se muestra una gráfica realizada usando la interfaz de Pandas, para lo que no tenemos más que hacer uso del parámetro **ax**:

```python
fig, ax = plt.subplots(2, 2, figsize = (12, 8))
flights.plot.line(
    x = "year",
    y = "passengers",
    marker = 'o',
    title = "Passengers by year",
    ylabel = "Passengers",
    grid = True,
    legend = False,
    ax = ax[0, 0]
)
flights.plot.bar(
    x = "year",
    y = "passengers",
    color = "salmon",
    legend = False,
    title = "Passengers by year",
    ylabel = "Passengers",
    grid = True,
    ax = ax[0, 1]
)
total_bill_by_day.plot.pie(
    y = total_bill_by_day,
    labels = total_bill_by_day.index,
    autopct='%1.1f%%',
    title = "Total bill by day",
    ylabel = "",
    ax = ax[1, 0]
)
tips.plot.hexbin(
    x = "total_bill",
    y = "tip",
    gridsize = (20, 12),
    title = "Total bill",
    xlabel = "Total bill",
    ylabel = "Tip",
    color = "lightgray",
    ax = ax[1, 1]
)
plt.tight_layout(pad = 1.0)
plt.show()
```

# La sublibrería pandas.plotting

La sublibrería **pandas.plotting** proporciona herramientas para facilitar la visualización de datos y la creación de gráficos complejos que van más allá de los gráficos básicos disponibles directamente a través del método *.plot()* en Series y DataFrames. Estas herramientas son particularmente útiles para la exploración de datos y la visualización de relaciones complejas dentro de los conjuntos de datos.

Veamos algunos de estas funciones:

### scatter_matrix()

La función [pandas.plotting.scatter_mattrix](https://pandas.pydata.org/docs/reference/api/pandas.plotting.scatter_matrix.html) muestra una matriz de gráficos de dispersión cruzando las características cuantitativas del dataframe indicado:

```python
pd.plotting.scatter_matrix(iris, figsize = (10, 10));
```

### parallel_coordinates()

La función [pandas.plotting.parallel_coordinates](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.plotting.parallel_coordinates.html) implementa el gráfico de [coordenadas paralelas](https://en.wikipedia.org/wiki/Parallel_coordinates), útil para trazar datos numéricos multivariados y comparar variables y sus relaciones.

En estos gráficos, cada característica recibe su propio eje, colocados verticalmente, y las muestras se representan como líneas entre los ejes:

```python
fig = plt.figure(figsize = (7, 7))
pd.plotting.parallel_coordinates(iris, "species");
```

### andrews_curves()

Las *curvas de Andrews*, implementadas en la función [pandas.plotting.andrews_curves](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.plotting.andrews_curves.html) son especialmente útiles para identificar si hay patrones distintivos que separan las categorías basados en las características de los datos. En muchos casos, las curvas de diferentes categorías mostrarán formas claramente distintas, lo que indica que las categorías son separables en el espacio multidimensional de sus características.

```python
plt.figure(figsize=(12, 6))
pd.plotting.andrews_curves(iris, "species")
plt.title("Andrews curves of Iris dataset")
plt.show()
```

### autocorrelation_plot()

La función [pandas.plotting.autocorrelation_plot](https://pandas.pydata.org/docs/reference/api/pandas.plotting.autocorrelation_plot.html) muestra la autocorrelación de una serie temporal para un cierto número de "*lags*", lo que puede ayudar a identificar si la serie temporal muestra alguna periodicidad o estacionalidad. Por ejemplo, si hay picos significativos a intervalos regulares en el gráfico, esto podría indicar un patrón estacional en los datos.

Para probar esta función, obtengamos el número de pasajeros transportados por mes a partir del dataset *flights* de Seaborn:

```python
data = sns.load_dataset("flights")
data.sort_values(by = ["year", "month"], inplace = True)
data.head(2)
```

```python
plt.figure(figsize=(10, 6))
pd.plotting.autocorrelation_plot(data["passengers"])
plt.title("Autocorrelation of monthly airline passengers over time")
plt.show()
```

### lag_plot()

El tipo de gráfico implementado en [pandas.plotting.lag_plot](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.plotting.lag_plot.html) es particularmente útil para identificar patrones o estructuras temporales en los datos, como tendencias o ciclicidad. Si los puntos en el gráfico forman una estructura claramente definida, esto indica una fuerte autocorrelación en la serie temporal, lo que podría deberse a patrones estacionales.

Podemos probarla en el dataset *flights*, aunque nos convendrá cargar el dataframe original:

```python
data = sns.load_dataset("flights")
```

```python
plt.figure(figsize=(8, 6))
pd.plotting.lag_plot(data["passengers"])
plt.title("Lag plot of airline passengers")
plt.show()
```

### bootstrap_plot()

El objetivo de la función [pandas.plotting.bootstrap_plot](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.plotting.bootstrap_plot.html) es visualizar la variabilidad de estimaciones estadísticas como la media, la mediana o los cuartiles mediante el uso de muestras bootstrap. Este tipo de gráfico puede ser especialmente útil para entender la incertidumbre en las estimaciones derivadas de muestras pequeñas.

```python
pd.plotting.bootstrap_plot(
    tips["total_bill"],
    size = 50,
    samples = 500,
    color = "Chocolate"
)
plt.show()
```

### radviz()

El algoritmo **RadViz** permite proyectar un conjunto de datos multidimensional en un espacio de dos dimensiones y la función [pandas.plotting.radviz](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.plotting.radviz.html) implementa este algoritmo, mostrando una gráfica de dos dimensiones con información del dataframe incluido como argumento.

Las características del dataframe son representadas distribuidas uniformemente a lo largo de la circunferencia de un círculo. Cada muestra del dataframe se representa en el interior del círculo de acuerdo con el valor en cada serie según una metáfora física: cada punto se supone unido a cada característica con una fuerza que es proporcional al valor que dicha muestra toma en la serie correspondiente, de forma que la posición final es la posición de equilibrio entre todas las fuerzas que representan las características.

Puedes acceder al artículo original aquí:

https://arxiv.org/pdf/2110.09971.pdf

Veamos qué resultado ofrece esta función aplicada al dataset *iris*:

```python
fig, ax = plt.subplots(figsize = (7, 7))
pd.plotting.radviz(iris, "species", ax = ax);
```

Intuitivamente podemos ver que la característica "*sepal_width*" toma valores proporcionalmente más elevados en la especie setosa que en versicolor o en virginica, pues los puntos correspondientes a este tipo de flor se muestran más próximos a esta característica que a las otras dos. Podemos confirmar este hecho fácilmente:

```python
iris.groupby("species")["sepal_width"].mean()
```
