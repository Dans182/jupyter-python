Algo que no hemos visto todavía -y que resulta extremadamente útil- es la posibilidad de pedir ayuda a Python sobre el uso de alguna función o método. Y tenemos varios métodos para esto: el primero es ejecutar una instrucción formada por el nombre del elemento sobre el que queremos pedir ayuda seguido de una interrogación:

En función de la versión de Jupyter que estemos usando veremos que se muestra la ayuda relativa a la función como resultado de ejecutar la celda:

O que se abre una ventana en la parte inferior del navegador con la ayuda solicitada:

Esta ventana puede ser maximizada o cerrada usando los controles que se muestran en su esquina superior derecha.

Una segunda opción es utilizar la función integrada **help** (que ya ha aparecido cuando hablamos de documentar nuestras funciones), incluyendo como argumento el nombre de aquello sobre lo que deseamos recibir ayuda:

```python
help(len)
```

Se ha escogido para los dos ejemplos la función *len* porque es muy simple de usar y su ayuda es muy escueta. Pero para la mayor parte de las funciones los textos de ayuda son muy completos (y largos). Si es el caso, te recomiendo usar la primera opción (nombre del elemento seguido de un signo de interrogación) y maximizar la ventana.

Si se ejecuta la instrucción help() sin argumentos, se abre una sesión de ayuda en la que se muestra una casilla en la que poder escribir el nombre de aquello sobre lo que se desea recibir ayuda, tal y como se muestra en la siguiente captura de pantalla:

<div style="font-size: 40px;text-align: center;height:50px;padding:10px;margin:0 0 10px 0;">Comentarios</div>

Es posible añadir comentarios a nuestro código. Éstos van precedidos por el símbolo "#". Cada vez que Python, a medida que ejecuta nuestro código línea por línea, se encuentre uno de estos símbolos, sabrá que desde el símbolo hasta el final de esa línea solo se trata de un comentario y simplemente ignorará dicho texto:

```python
a = 1     # Mi variable a
b = 2     # Y mi variable b
```

Si queremos añadir comentarios que ocupen varias líneas podemos recurrir a añadir la almohadilla al comienzo de cada una:

```python
a = 1
# Esto es un
# comentario que
# ocupa varias
# líneas
b = 2
```

O podemos usar los bloques de triples comillas dobles (o sencillas) que ya conocemos:

```python
a = 1
"""
Esto es
un comentario que
ocupa varias
líneas
"""
b = 2
```
