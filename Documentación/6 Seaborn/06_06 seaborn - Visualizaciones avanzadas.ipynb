```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
flights = sns.load_dataset("flights")
iris = sns.load_dataset("iris")
```

# Mapas de calor

Seaborn ofrece la función a nivel de ejes [seaborn.heatmap](https://seaborn.pydata.org/generated/seaborn.heatmap.html) para la generación de mapas de calor. En su forma más simple, tan solo requiere como argumento de entrada un bloque de valores de dos dimensiones, valores que serán representados por la paleta de colores escogida. Si este bloque es un dataframe, las etiquetas de filas y columnas se mostrarán como etiquetas en el mapa de calor.

Para ver un ejemplo, podemos basarlos en el dataset *flights*, a partir del cual podemos generar una estructura bidimensional que cruce meses y años, conteniendo el número de pasajeros:

```python
data = pd.pivot_table(data = flights, index = "month", columns = "year", values = "passengers")
data
```

Ahora basta pasar esta estructura a la función:

```python
ax = sns.heatmap(data)
plt.show()
```

Podemos especificar un mapa de color con el parámetro **cmap**, al que podemos asignar el nombre de una paleta de colores:

```python
ax = sns.heatmap(data, cmap = "Blues")
plt.show()
```

También podemos mostrar los valores en las casillas con el parámetro **annot**:

```python
ax = sns.heatmap(data, cmap = "Blues", annot = True)
plt.show()
```

# Mapa de clusters

La función [seaborn.clustermap](https://seaborn.pydata.org/generated/seaborn.clustermap.html) es útil para crear un mapa de calor que reorganiza los datos para mostrar patrones mediante un algoritmo de agrupación jerárquica (clustering).

Probémosla con el DataFrame *data* que hemos creado que contiene el número de pasajeros por año y mes:

```python
sns.clustermap(data, cmap="coolwarm", standard_scale=1)
plt.show()
```

# Pairplot

Ya conocemos la función [seaborn.pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html), función a nivel de figura que crea una matriz de ejes y muestra la distribución de cada par de características:

```python
sns.pairplot(iris, hue = "species")
plt.show()
```

# Pairgrid

La función [seaborn.PairGrid](https://seaborn.pydata.org/generated/seaborn.PairGrid.html) es semejant a *sns.pairplot()* pero, a diferencia de ésta, nos da un control total sobre los gráficos a mostrar en cada parte de la matriz de gráficos, permitiendo una personalización completa:

```python
g = sns.PairGrid(iris, hue = "species")

g.map_upper(sns.scatterplot, edgecolor="w")  # Gráficos de dispersión por encima de la diagonal
g.map_diag(sns.histplot)                     # Histogramas en la diagonal
g.map_lower(sns.kdeplot, levels = 4)         # Gráficos de contornos KDE por debajo de la diagonal

g.add_legend()
plt.show()
```
