# Colores

El color es, sin duda, la variable visual más importante, y su correcta elección puede determinar la eficacia de una visualización. Seaborn incluye varias funciones para el uso y gestión de paletas de colores, siendo las más destacadas [seaborn.color_palette](https://seaborn.pydata.org/generated/seaborn.color_palette.html) y [seaborn.set_palette](https://seaborn.pydata.org/generated/seaborn.set_palette.html). Revisemos éstas y otras funciones relacionadas.

### Paletas discretas

La función más importante para trabajar con colores discretos es [seaborn.color_palette](https://seaborn.pydata.org/generated/seaborn.color_palette.html). Esta función devuelve la lista de colores representada por el nombre de la paleta que se indique. Por ejemplo:

```python
sns.color_palette("husl")
```

En este caso, la paleta "*husl*" está compuesta por 6 colores, representado cada uno por tres valores correspondientes a los canales rojo, verde y azul, colores que son los mostrados como resultado de la ejecución del código anterior.

Podemos visualizar los componentes rojo, verde y azul de cada uno de ellos:

```python
[color for color in sns.color_palette("husl")]
```

Esta función acepta cualquier nombre de paleta definida en Matplotlib o en Seaborn.

También acepta un número entero que indicará el número de colores a generar (si este número es mayor que el número de colores disponibles en la paleta, se repetirán cíclicamente).

Muchas de las funciones de visualización que hemos visto aceptan el parámetro **palette**, al que podemos asignar un listado de colores como el devuelto por la función *sns.color_palettes*:

```python
sns.catplot(data = tips, x = "day", y = "total_bill", kind = "violin", palette = "husl")
plt.show()
```

```python
sns.catplot(data = titanic, x = "deck", kind = "count", palette = "Set2")
plt.show()
```

Cuando trabajemos con variables categóricas entre cuyos valores no sea posible establecer una relación de orden, es importante seleccionar colores que comuniquen esa independencia de valores, evitando, por ejemplo, gradaciones, o colores que puedan transmitir una idea de dependencia entre las distintas categorías de valores.

Podemos consultar un listado de paletas de Matplotlib en la siguiente página de su documentación:

https://matplotlib.org/stable/users/explain/colors/colormaps.html

Sorprendentemente, no hay forma de consultar el listado completo de paletas ofrecidas por Seaborn. En la siguiene página web de su documentación se muestran algunas (sin especificarse si hay o no otras):

https://seaborn.pydata.org/tutorial/color_palettes.html

### Paletas secuenciales

Este tipo de paletas son apropiadas cuando existe un orden o gradación en los datos. En Seaborn tenemos numerosas paletas de este tipo. Algunas llevan el nombre del color predominante:

```python
sns.color_palette("Blues")
```

```python
sns.color_palette("Reds")
```

```python
sns.color_palette("Greens")
```

Si preferimos que una paleta se muestre invertida, tal y como ocurre en matplotlib, basta con añadir "*_r*" al final del nombre:

```python
sns.color_palette("Greens_r")
```

Seaborn incorpora un pequeño truco consistente en añadir el sufijo "*_d*" (*dark*) para conseguir paletas más oscuras. Comparemos, por ejemplo, la siguiente paleta que va de un verde a un azul:

```python
sns.color_palette("GnBu")
```

con la correspondiente "oscura":

```python
sns.color_palette("GnBu_d")
```

### Paletas divergentes

El tercer tipo de paletas es el llamado divergente. Se usan en escenarios en los que hay un orden entre los valores pero hay también un "centro" definido. Por ejemplo, alturas con respecto al nivel del mar (que pueden ser positivas, negativas o cero):

```python
sns.color_palette("BrBG", 11)
```

```python
sns.color_palette("coolwarm", 11)
```

### Paleta de colores por defecto

Es posible establecer una paleta de colores por defecto con la función [seaborn.set_palette](https://seaborn.pydata.org/generated/seaborn.set_palette.html). Una vez ejecutada, la paleta en cuestión se aplicará a todas las visualizaciones que se creen a partir de ese momento:

```python
sns.set_palette("colorblind")
```

```python
sns.catplot(data = tips, x = "day", y = "total_bill", hue = "sex", kind = "swarm")
plt.show()
```

```python
sns.set_palette("cubehelix")
```

```python
sns.pairplot(iris, hue = "species")
plt.show()
```

### Paletas personalizadas

Tal y como hemos visto, una paleta consiste en una lista de valores correspondientes a los colores de la paleta. Estos valores pueden venir dados por tuplas de tres valores -cada uno definiendo el valor del canal rojo, verde y azul (RGB)- por su código hexadecimal o incluso por el nombre del color.

Esto nos permite crear nuestras propias paletas -por ejemplo con los colores corporativos de nuestra empresa- y aplicarlas a las visualizaciones:

```python
my_palette = ["#D4E09B", "#9AC2C9", "#9CB380", "#C2C5BB"]
```

```python
sns.catplot(data = tips, x = "tip", y = "day", kind = "box", palette = my_palette)
plt.show()
```

# Estilos

Otra de las herramientas de las que disponemos para controlar el aspecto gráfico de nuestras visualizaciones son los estilos. Éstos definen el grid, la fuente a utillizar, las marcas de los ejes, etc.

### Estilos disponibles

Seaborn incluye 5 estilos disponibles: "*darkgrid*", "*whitegrid*", "*dark*", "*white*" y "*ticks*". La función [seaborn.set_style](https://seaborn.pydata.org/generated/seaborn.set_style.html) nos permite establecer uno como activo:

```python
sns.set_palette("tab10")
for i, s in enumerate(["dark", "white", "ticks", "whitegrid", "darkgrid"]):
    sns.set_style(s);
    fig = plt.figure(figsize = (6, 2))
    fig.suptitle(s)
    sns.boxplot(data = tips, x = "day", y = "tip")
plt.show()
```

### Componentes de un estilo

Un estilo viene definido por un diccionario en el que se asignan los valores adecuados a cada uno de los elementos susceptibles de ser modificados. Una vez fijado un estilo podemos ver sus componentes utilizando la función [sns.axes_style](https://seaborn.pydata.org/generated/seaborn.axes_style.html):

```python
sns.set_style("white")
sns.axes_style()
```

### Uso de estilos modificados

Es posible establecer como activo un estilo modificando alguno de sus elementos. Para ello no hay más que utilizar la función ya vista [sns.set_style](https://seaborn.pydata.org/generated/seaborn.set_style.html) incluyendo como segundo argumento (tras el nombre del estilo a usar) un diccionario en el que se fijen los valores que se deseen cambiar.

Veamos un ejemplo. Partimos del estilo "*white*":

```python
sns.set_style("white")
sns.boxplot(data = tips, x = "day", y = "tip")
plt.show()
```

Supongamos que queremos mostrar este mismo gráfico pero incluyendo el grid y mostrando sus líneas de color azul claro. Podríamos conseguirlo de la siguiente forma:

```python
sns.set_style("white", {"axes.grid": True, "grid.color": "lightblue"})
sns.boxplot(data = tips, x = "day", y = "tip")
plt.show()
```

### Creación de estilos

Y, sabiendo que un estilo es un diccionario, podemos también crear nuestros propios estilos:

```python
myStyle = {
    'axes.edgecolor': '#333333',
    'axes.facecolor': '#AAAAAA',
    'axes.grid': True,
    'grid.linestyle': '--',
    "grid.color": "white",
    'ytick.right': True,
    'axes.spines.right': False,
    'axes.spines.top': False
}
```

Y fijarlos como activos:

```python
sns.set_style("white")
sns.set_style(myStyle)
sns.boxplot(data = tips, x = "day", y = "tip")
plt.show()
```

Los elementos del estilo a los que no demos formato se mantienen tal y como estén en ese momento. Es por ello que, en el ejemplo anterior, primero se fija el estilo "white" y después el nuestro.
