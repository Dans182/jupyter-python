# Presentación

Seaborn es una librería de visualización de datos para Python desarrollada sobre Matplotlib. Ofrece una interfaz de alto nivel para la creación de atractivas gráficas. Además, está íntimamente integrada con las estructuras de datos de Pandas, lo que permite utilizar el nombre de los DataFrames y campos directamente como argumentos de las funciones de visualización.

Seaborn tiene como objetivo convertir la visualización en una parte central de la exploración y comprensión de los datos, permitiendo generar gráficas con sencillas funciones que ofrecen una interfaz semejante, facilitando el paso de unas funciones a otras.

La documentación oficial está disponible en el sitio web oficial de Seaborn:

https://seaborn.pydata.org/

Esta librería se importa habitualmente con el alias **sns**:

```python
import seaborn as sns
```

Importemos también otras librerías que podemos necesitar:

```python
import matplotlib.pyplot as plt
import numpy as np
import warnings
```

```python
warnings.filterwarnings("ignore")
```

Y carguemos un DataFrame Pandas proveído por la propia librería Seaborn:

```python
tips = sns.load_dataset("tips")
```

# Funcionalidad de Seaborn

Seaborn ofrece varias características destacadas:

* Funciones con una API estandarizada que limita la necesidad de tener que memorizar conjuntos de parámetros diferentes en función del gráfico deseado, como en las tres siguiente gráficas, para cuya generación solo es necesario modificar el valor de un parámetro de la función correspondiente:

* Gráficas visualmente atractivas sin necesidad de realizar complicados ajustes



* Una API orientada a conjuntos de datos para examinar la relación entre múltiples variables.



* Opciones para mostrar la distribución de variables univariadas y bivariadas



* Cálculo automático y dibujo de modelos de regresión lineal para diferentes tipos de variables dependientes



* Herramientas para mostrar la estructura de datasets complejos



* Abstracciones de alto nivel para estructurar rejillas multigráficas con el objetivo de poder crear complejas visualizaciones



* Sencillo control sobre los estilos gráficos disponibles
* Herramientas para la elección de paletas de color adecuadas que permitan revelar patrones en los datos

# Funciones a nivel de figura y de ejes

En Seaborn hay dos tipos de funciones gráficas: funciones a nivel de figura y funciones a nivel de ejes. Las primeras son funciones de alto nivel que gestionan la figura y el conjunto de ejes que las contiene, así como la leyenda (recordemos que Seaborn está construida sobre Matplotlib, por lo que, cuando ejecutamos funciones de Seaborn, en realidad estamos ejecutando funciones de Matplotlib). Funciones de este tipo son [seaborn.relplot](https://seaborn.pydata.org/generated/seaborn.relplot.html) y [seaborn.catplot](https://seaborn.pydata.org/generated/seaborn.catplot.html), por ejemplo.

Las funciones a nivel de ejes, por el contrario, se van a crear en la figura y conjunto de ejes que estén activos, o en aquel conjunto de ejes que se indique como argumento.

Algunas de las funciones a nivel de figura tienen equivalentes a nivel de ejes. Por ejemplo, con la función a nivel de figura *sns.catplot()* podemos mostrar un gráfico de caja, o un gráfico de violín, entre otros. Pues bien, existen también las funciones a nivel de ejes *sns.boxplot()* y *sns.violinplot()* que muestran también gráficos de caja y de violín, respectivamente. Para verlo en funcionamiento ejecutemos la función a nivel de figura *sns.catplot()* indicando mediante el parámetro **kind** qué tipo de gráfica queremos generar:

```python
sns.catplot(data = tips, x = "day", y = "total_bill", kind = "swarm")
plt.show()
```

Pero, si en lugar de la anterior instrucción, ejecutamos la función a nivel de ejes *seaborn.swarmplot()*:

```python
sns.swarmplot(data = tips, x = "day", y = "total_bill")
plt.show()
```

el resultado es semejante al anterior, pero quedando en nuestras manos el control de la figura y del conjunto de ejes en el que se va a crear.

Otro ejemplo: la función a nivel de figura [seaborn.relplot](https://seaborn.pydata.org/generated/seaborn.relplot.html) puede mostrar una gráfica de dispersión (si el parámetro *kind* toma el valor "*scatter*", que es su valor por defecto), al igual que la función a nivel de ejes [seaborn.scatterplot](https://seaborn.pydata.org/generated/seaborn.scatterplot.html), o un gráfico de líneas (si el parámetro *kind* toma el valor "*line*"), igual que la función a nivel de ejes [seaborn.lineplot](https://seaborn.pydata.org/generated/seaborn.lineplot.html). Veremos estas funciones en breve.

A la hora de crear visualizaciones podemos utilizar tanto un tipo como el otro. Tan solo deberemos saber que no podremos controlar la figura ni el conjunto de ejes de las funciones a nivel de figura, cosa que sí podremos hacer con las funciones a nivel de eje. Posiblemente las funciones a nivel de figura son más apropiadas para análisis exploratorio, pues no tenemos que molestarnos en crear la figura, la leyenda, etc.

Estrictamente hablando, sí podemos controlar características de una función a nivel de figura (figura, ejes...), pero solo interacturando con la estructura de clases de Matplotlib.

Podemos distinguir unas funciones de otras viendo si admiten el parámetro **ax**: Si lo admiten, se trata de funciones a nivel de ejes. Otra forma de saberlo es viendo el tipo del resultado que devuelven: si se trata de un objeto tipo "*axes*" de Matplotlib, se trata de una función a nivel de ejes. Si, por el contrario, se trata de un objeto "*FacetGrid*", es una función a nivel de figura. Otra diferencia es que la mayor parte de las funciones a nivel de figura (no todas) muestran la legenda fuera del gráfico, mientras que las funciones a nivel de ejes la incluyen dentro del gráfico.

# Datasets

Seaborn incluye un interesante conjunto de datasets que pueden ser utilizados para probar las distintas visualizaciones. Podemos visualizar el listado de datasets disponibles:

```python
print(sns.get_dataset_names())
```

Para cargar un dataset podemos usar la función [seaborn.load_dataset](https://seaborn.pydata.org/generated/seaborn.load_dataset.html), devolviéndose un dataframe Pandas. Revisemos algunos de estos datasets:

#### flights

Dataset con información sobre pasajeros transportados en avión por mes entre 1949 y 1960. Las cifras de pasajeros probablemente hacen referencia a "miles" de pasajeros (no está documentado):

```python
flights = sns.load_dataset("flights")
flights.sample(5)
```

#### titanic

Famoso conjunto de datos con información sobre los pasajeros del último viaje del Titanic, con información sobre quiénes sobrevivieron, su edad, sexo, clase en la que viajaban...:

```python
titanic = sns.load_dataset("titanic")
titanic.sample(5)
```

#### iris

Otro de los clásicos, con información de la longitud y ancho de pétalos y sépalos de 150 flores Iris de tres especies distintas:

```python
iris = sns.load_dataset("iris")
iris.sample(5)
```

#### tips

En este conjunto de datos hay información sobre propinas dejadas en un restaurante, con información sobre el día en el que se produjo, si fue en el almuerzo o la cena, el número de comensales, el sexo de la persona que dejó la propina, si era fumador o no...:

```python
tips = sns.load_dataset("tips")
tips.sample(5)
```

#### fmri

Datos obtenidos de imágenes por resonancia magnética funcional:

```python
fmri = sns.load_dataset("fmri")
fmri.sample(5)
```
